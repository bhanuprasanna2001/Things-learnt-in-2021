```python
from IPython.display import clear_output

print('Welcome to Tic Tac Toe\n**********************')

def display_b():
    print('   |   |   ')
    print(' 1 | 2 | 3 ')
    print('_ _|_ _|_ _')
    print('   |   |   ')
    print(' 4 | 5 | 6 ')
    print('_ _|_ _|_ _')
    print('   |   |   ')
    print(' 7 | 8 | 9 ')
    print('   |   |   ')

def place_marker(board, marker, position):
    board[position] = marker

def create_board():
    return [' '] * 10

def disb(b):
    print('   |   |   ')
    print(f' {b[1]} | {b[2]} | {b[3]}')
    print('_ _|_ _|_ _')
    print('   |   |   ')
    print(f' {b[4]} | {b[5]} | {b[6]}')
    print('_ _|_ _|_ _')
    print('   |   |   ')
    print(f' {b[7]} | {b[8]} | {b[9]}')
    print('   |   |   ')

def player_choice(board):
    position = 0
    
    while position not in [1,2,3,4,5,6,7,8,9] or not check_space(board, position):
        position = int(input('Choose your next position: (1-9) '))
        
    return position

def replay():
    return input("Do you want to play the game again (yes, no) : ").lower().startswith('y')

def check_space(b, pos):
    return b[pos] == ' '

def check_full(b):
    for i in range(1,10):
        if b[i] == ' ':
            return False
    return True

def check_win(b, mark):
    return  ((b[1] == mark and b[2] == mark and b[3] == mark) or # Horizontal 1
            (b[4] == mark and b[5] == mark and b[6] == mark) or  # Horizontal 2
            (b[7] == mark and b[8] == mark and b[9] == mark) or  # Horizontal 3
            (b[1] == mark and b[4] == mark and b[7] == mark) or  # Vertical 1
            (b[2] == mark and b[5] == mark and b[8] == mark) or  # Vertical 2
            (b[3] == mark and b[6] == mark and b[9] == mark) or  # Vertical 3
            (b[1] == mark and b[5] == mark and b[9] == mark) or  # Diagonal 1
            (b[3] == mark and b[5] == mark and b[7] == mark))    # Diagonal 2


while True:
    b = create_board()
    player1 = input('Enter (X, O): ').upper()
    while player1 not in ['X', 'O']:
        player1 = input('Enter (X, O): ').upper()
    player2 = 'X' if player1 == 'O' else 'O'
    
    chance = '1' if player1 == 'X' else '2'
    game_on = True
    print(player1 + player2)
    
    while game_on:
        # Player 1
        if chance == '1':
            display_b()
            disb(b)
            choice = player_choice(b)
            place_marker(b, player1, choice)
            clear_output()
            
            if check_win(b, player1):
                clear_output()
                disb(b)
                print(f'Player {chance}, you have won.')
                game_on = False
            else:
                if check_full(b):
                    clear_output()
                    disb(b)
                    print('Tie!')
                    break
                else:
                    chance = '2'
        else:
            display_b()
            disb(b)
            choice = player_choice(b)
            place_marker(b, player2, choice)
            clear_output()
            
            if check_win(b, player2):
                clear_output()
                disb(b)
                print(f'Player {chance}, you have won.')
                game_on = False
            else:
                if check_full(b):
                    clear_output()
                    disb(b)
                    print('Tie!')
                    break
                else:
                    chance = '1'
    if not replay():
        break
```
